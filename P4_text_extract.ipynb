```python
from google.colab import userdata
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import pytesseract
import easyocr

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'


import warnings
warnings.filterwarnings('ignore')
```

```python
# load files from google drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
image_dir = '/content/drive/MyDrive/Colab Notebooks/Apziva/MonReader/books'
image_files = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith(('.jpg'))]

plt.figure(figsize=(15, 10))
for i, img_path in enumerate(image_files):
  if i < 10: # Limit the number of images shown
    img = plt.imread(img_path)
    plt.subplot(2, 5, i + 1)
    plt.imshow(img)
    plt.title(os.path.basename(img_path))
    plt.axis('off')
  else:
    break
plt.tight_layout()
plt.show()
```

## Pytesseract

```python
page3_tes = pytesseract.image_to_string(image_files[0], lang='eng')
print(page3_tes)
```

```python
page4_5_tes = pytesseract.image_to_string(image_files[1], lang='eng')
print(page4_5_tes)
```

## easyocer

```python
reader = easyocr.Reader(['en'], gpu=True)
```

```python
result = reader.readtext(image_files[0])
for (bbox, page3, prob) in result:
  print(page3)
```

```python
result = reader.readtext(image_files[1])
for (bbox, page4_5, prob) in result:
  print(page4_5)
```
